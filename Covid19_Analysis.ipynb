```python
data.groupby('Region')[['Confirmed','Recovered','Deaths']].sum()
```

# Q2. Remove all the records where Confirmed case is les than 10.

```python
data.head()
```

```python
data=data[~(data['Confirmed']<=10)]
```

```python
data.head(20)
```

# Q3. In which region, maximum number of Confirmed cases were recorded?

```python
data.head()
```

```python
data.groupby('Region')['Confirmed'].sum().sort_values(ascending=False)
```

# Q4. Provide the top 5 region, minimum number of Deaths cases were recorded?

```python
data.groupby('Region')['Deaths'].min().sort_values(ascending=False).nlargest(5)
```

```python
# Now visualising

plt.title('Top 5 Regions with Minimum Deaths')
data.groupby('Region')['Deaths'].min().sort_values(ascending=False).nlargest(5).plot(kind='bar')
plt.xticks(rotation=0)
plt.xlabel('Region')
plt.ylabel('Deaths')
plt.savefig('barchart.png')
plt.show()
```
